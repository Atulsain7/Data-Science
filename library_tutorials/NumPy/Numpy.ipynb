### Question 4:

The power emitted by a blackbody is $P=A\sigma\epsilon T^{4}$. After measuring the temperature of a star you find that $T(t)=T_{o}\frac{1}{\epsilon^{kt}}$.<br>
Plot the total enegy emitted by the star as a fuction of time using the fact that $E(t)=\int_{t'=0}^{t'=t}P(t')dt'$.

$$ New\_variable\ (kt)=\frac{P}{A\sigma \epsilon T_{o}^4}=\Big(\frac{1}{1+e^{-kt}}\Big)^{4}$$

```python
kt=np.linspace(0,3,100)
P=(1/(1+np.exp(-kt)))**4
E=np.cumsum(P)*(kt[1]-kt[0])
```

```python
plt.plot(kt,P,label='$kt$ Power function')
plt.plot(kt,E,label='$kt$ Energy fuction')
plt.xlabel("$kt$")
plt.ylabel(r"$kt=\left(\frac{P}{A \sigma \epsilon T_o^{4}}\right) , E(kt)$",fontsize=15)
plt.legend()
```

### Question 3:
An electric field is given by $\vec{E}(x,t)=E_{o}\cos{(z-t)}\hat{x}+2E_{o}\cos{(z-t+\pi /2)}\hat{y}$<br><br>
1.Find the magnetic field for all $z\in [0,4\pi]$ and $t\in [0,10]$ using $\hat{B}(z,t)=\hat{z}\times \vec{E}(z,t)$<br><br>
2.Compute the Poynting vector $\vec{S}=\vec{E}\times \vec{B}$ for all $z$ and $t$.

```python
#1
z=np.linspace(0,4*np.pi,100)
t=np.linspace(0,10,100)

tv,zv=np.meshgrid(t,z)

Ex=np.cos(zv-tv) #its Ex/Eo
Ey=2*np.cos(zv-tv+np.pi/2)
Ez=0*tv
```

$E_z$ as function of $z$ at $t=0$

```python
plt.plot(t,Ex[0],label='Ex')
plt.plot(t,Ey[0],label='Ey')
plt.xlabel('t')
plt.legend()
```

$E_x$ as funtion of $t$ at $z=0$

```python
plt.plot(Ex[:,0])
```

```python
E = np.array([Ex, Ey, Ez])
E = np.swapaxes(E, 0, -1)
```

```python
B = np.cross(np.array([0,0,1]), E)
E = np.swapaxes(E, 0, -1)
B = np.swapaxes(B, 0, -1)
```

```python
Bx, By, Bz = B
```

```python
plt.plot(t,Ey[0])
plt.plot(t,Bx[0])
```
